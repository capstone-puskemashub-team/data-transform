```python
dt = df[['Umur','Keluhan Utama', 'Suhu(C`)', 'Tekanan Darah (mmHg)', 'Nadi (x/menit)',
       'Berat Badan(Kg)', 'Tinggi Badan(cm)', 'Diagnosis']]

dt.rename(columns={
    'Umur':'umur',
    'Keluhan Utama':'keluhan_utama',
    'Suhu(C`)':'suhu', 
    'Tekanan Darah (mmHg)':'tkd', 
    'Nadi (x/menit)':'nadi',
    'Berat Badan(Kg)':'bb', 
    'Tinggi Badan(cm)':'tb'
}, inplace=True)
dt.head()
```
